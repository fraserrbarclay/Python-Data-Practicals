```python
def split_column(df,column,q_intro):
    for response in df[column].unique():
      response_column = f"{q_intro}:{response}"
      df[response_column] = 'No answer'
      df.insert(df.columns.get_loc(column), response_column, df.pop(response_column))

# Iterate through the rows and update 'Yes' for matching response columns
    for index, row in df.iterrows():
      response = row[column]
      response_column_name = f"{q_intro}:{response}"
      df.at[index, response_column_name] = 'Yes'

    df.drop(columns=[column], inplace=True)
    empty_column = f"{q_intro}:nan"
    if empty_column in df:
      df.drop(columns=[empty_column], inplace=True)
    return df
```
