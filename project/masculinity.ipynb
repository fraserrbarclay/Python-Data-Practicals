```python
def split_column(df,column,q_intro):
    for response in df[column].unique():
      response_column = f"{q_intro}:{response}"
      df[response_column] = 'No answer'
      df.insert(df.columns.get_loc(column), response_column, df.pop(response_column))

# Iterate through the rows and update 'Yes' for matching response columns
    for index, row in df.iterrows():
      response = row[column]
      response_column_name = f"{q_intro}:{response}"
      df.at[index, response_column_name] = 'Yes'

    df.drop(columns=[column], inplace=True)
    empty_column = f"{q_intro}:nan"
    if empty_column in df:
      df.drop(columns=[empty_column], inplace=True)
    return df
```

```python
import numpy as np

def remove_missing(df) :
  for i in range (len(df.columns)):
    df.loc[df.iloc[:, i] == 'Not selected', df.columns[i]] = np.nan
    df.loc[df.iloc[:, i] == 'No answer', df.columns[i]] = np.nan
  return df
remove_missing(df)
```

```python
def normalise_yn(df) :
  for i in range (len(df.columns)):
    df.loc[df.iloc[:, i] == 'Yes', df.columns[i]] = 4
    df.loc[df.iloc[:, i] == 'No', df.columns[i]] = 0
    df.loc[df.iloc[:, i] == 'Often', df.columns[i]] = 4
    df.loc[df.iloc[:, i] == 'Sometimes', df.columns[i]] = 3
    df.loc[df.iloc[:, i] == 'Rarely', df.columns[i]] = 2
    df.loc[df.iloc[:, i] == 'Never, but open to it', df.columns[i]] = 1
    df.loc[df.iloc[:, i] == 'Never, and not open to it', df.columns[i]] =0
  return df
normalise_yn(df)
```
