```python
def collapse_columns(df, first, last, name):
  q = df.iloc[:,first:last] .stack().reset_index(drop=False)
  q = q[['level_0',0]]
  q = q.groupby('level_0')[0].apply(list)

  deleted = df.iloc[:,first:last].columns
  for i in deleted:
    del df [i]

  df.insert(first,name ,q)
  return df

df = collapse_columns(df, 2,8 , 'q0004')
```
