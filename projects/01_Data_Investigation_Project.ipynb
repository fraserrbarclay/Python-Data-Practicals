```python

import pandas as pd
url = "https://github.com/frbarclay/Python-Data-Practicals/raw/Data-Sets/housing_in_london_yearly_variables.csv"
df = pd.read_csv(url)
df
```

```python
#I describe what this dataframe does
headings = df.columns[3:11]
for head in headings:
  print ("This data describes the", head, "of different areas of the UK over time")
```

```python
areas = df['area'].unique().tolist()
print("There are", len(areas), "distinct areas described by this dataframe.")
print('')
df_boroughs = df[(df['borough_flag'] != 0)]
df_regions = df[(df['borough_flag'] == 0)]
boroughs = df_boroughs['area'].unique()
regions = df_regions['area'].unique()

print("There are", len(boroughs), "distinct 'boroughs' (including the City of London) described by this dataframe.:")
for borough in boroughs:
  print(borough)
print('')
print("There are" , len(regions),  "regional areas described, for comparison")
print('The areas described are:')
for region in regions:
  print (region)
```

```python
#I can add columns for population density and persons per house/job

for i in df:
  pop_density = df['population_size']/df['area_size']
  jobs_pp = df ['number_of_jobs']/df['population_size']
  person_house =df['population_size']/ df ['no_of_houses']
df['population_density'] = pop_density
df['job_density'] = jobs_pp
df['persons_per_house'] = person_house
```

```python
#The data range is already in date order
for date in df['date'].head(1):
  Date1 = date
for date in df['date'].tail(1):
  DateRecent= date
print('This data covers the time frame from',Date1,'to',DateRecent)
```

```python
#Every year will relate to every set of 51
#This function will return data depending on whichever year you'd like

def dfyear(year):
  if year > 2019 or year < 1999:
    return('No data available')
  else:
    startset = (year - 1999)*51
    endset = startset + 51
    return(df.iloc[startset:endset].drop(columns='code'))


df_year = dfyear(2009)
df_year
```
